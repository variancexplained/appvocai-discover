## Methods
The app reviews dataset was obtained from the Apple App Store over the course of two weeks in late August of 2023. The data were gathered from publicly accessible sources within the App Store, ensuring transparency and adherence to ethical guidelines throughout the process. 

### Data Acquisition

The data acquisition process for the AppVoC dataset employs a multi-layered system to efficiently collect, process, validate, and store app reviews. This system is composed of domain entities for reviews and scraping requests, application services for orchestrating the scraping and validation processes, infrastructure components for handling HTTP requests and rate limiting, and repository layers for managing data persistence and integrity.

```{figure} ../figures/AppVoCDataAcquisition.png
:name: data-acquisition-fig-ref

AppVoC Data Acquisition Architecture
```

#### Domain Layer
- **Review**: This class encapsulates individual app review details, including attributes like `id`, `app_id`, `app_name`, `category_id`, `category`, `author`, `rating`, `title`, `content`, `vote_sum`, `vote_count`, and `date`. Author anonymity is ensured using the BLAKE2b cryptographic hash function.
- **ReviewRequest**: Defines scraping requests with attributes like `id`, `category_id`, and `last_index` to track progress.
- **Job**: Represents an entity for job management with attributes such as `id`, `controller`, `category_id`, `category`, `complete`, and `completed`.
- **ReviewJobRun**: Models the execution of review jobs with attributes tracking app and review metrics.
- **ReviewResult**: Encapsulates app review data obtained from web scraping.

#### Application Layer
- **ReviewDirector**: Orchestrates job runs for the `ReviewController`, using a Unit of Work (UoW) pattern for repository interactions.
- **ReviewController**: Manages the scraping process, handling job runs, app retrieval, review data scraping, validation, and data persistence.
- **ReviewValidator**: Validates HTTP responses, ensuring the integrity and correctness of the review data.
- **ReviewScraper**: Extracts app reviews from the App Store, handling HTTP sessions, and pagination through multiple pages of reviews.

#### Infrastructure Layer
- **ASessionHandler**: Manages asynchronous HTTP requests, incorporating request retries, throttling, and proxy server integration.
- **AThrottle**: Implements asynchronous rate limiting by adjusting request delays based on real-time latency measurements.

#### Repository Layer
- **UoW (Unit of Work)**: Manages repositories for project objects within a structured database context.
- **ReviewRequestRepo**: Handles the management of review requests within a database.
- **ReviewRepo**: Stores and accesses app review data within a database context.
- **JobRepo**: Manages and tracks the progress of ETL jobs within a database.
- **ReviewJobRunRepo**: Manages review job runs within a database.

This architecture ensures systematic data management, robust process orchestration, and maintenance of data integrity across the entire workflow. It supports the efficient collection and processing of app reviews, enabling comprehensive analysis of user feedback and app interactions.

### Sampling Strategy

To ensure a representative and insightful analysis of app reviews, we established specific inclusion criteria for selecting categories. The criteria aim to balance diversity, user engagement, and functional relevance across various app types. Below are the articulated inclusion criteria based on the categories selected:

1. **High User Engagement and Relevance**:
   - Categories selected should represent areas with high user engagement and relevance to daily activities, professional tasks, and personal interests. This ensures that the insights derived are impactful and reflect significant user interactions.

2. **Diverse Functional Coverage**:
   - The chosen categories should cover a wide range of functions and utilities to provide a holistic view of app usage. This includes productivity, education, lifestyle, health, finance, entertainment, and information-related apps, ensuring a comprehensive understanding of user needs.

3. **Balanced Representation Across Major App Types**:
   - The selection should include categories that collectively represent a balanced cross-section of the app ecosystem. This means including high-percentage categories like Business, Education, and Utilities, as well as lower-percentage but critical categories like Medical and Social Networking.

4. **Potential for Impactful Insights**:
   - Categories that are likely to provide actionable insights for app development and improvement are prioritized. This includes areas where user feedback can directly inform enhancements in functionality, user experience, and engagement.

5. **Consistency in User Behavior and Engagement Patterns**:
   - Selected categories should exhibit relatively consistent and comparable user behavior and engagement patterns. This facilitates more accurate and meaningful analysis, avoiding the variability and trends often seen in highly volatile categories like gaming.

#### Justification for Excluding Gaming, the Largest Category

While games represent the largest category on the App Store with 12.68% of active apps, there are several strategic reasons for excluding it from the current sampling strategy:

1. **Niche Focus on Non-Entertainment Apps**:
   - The primary goal of this analysis is to understand user engagement and satisfaction across various functional and utility-based app categories. Including gaming, which is heavily entertainment-focused, could skew the results and dilute insights related to productivity, education, and lifestyle applications.

2. **Distinct User Behavior and Engagement Patterns**:
   - Gaming apps often exhibit unique user behavior and engagement patterns that differ significantly from other app categories. By excluding games, we can ensure a more consistent and comparable analysis across the selected categories, focusing on practical and utilitarian app usage.

3. **High Volatility and Trends in Gaming**:
   - The gaming industry is characterized by rapid changes in trends and high volatility, with popular games frequently emerging and declining. This can introduce variability and noise into the data, making it challenging to derive stable and long-term insights from the sample.

4. **Prioritizing Diversity and Depth**:
   - Given the limited scope of the study, prioritizing a diverse range of non-gaming categories allows for a deeper exploration of various aspects of user life and needs. This diversity ensures a more comprehensive understanding of app usage beyond entertainment.

5. **Overrepresentation Concerns**:
   - Including the largest category like gaming could lead to its overrepresentation, potentially overshadowing insights from smaller but equally important categories. Focusing on a balanced selection of categories ensures that each area receives adequate attention in the analysis.

6. **Alignment with Strategic Goals**:
   - The current study aims to gather insights that can inform the development and improvement of non-entertainment apps. By concentrating on categories like Business, Education, Utilities, and Health & Fitness, the findings can more effectively guide strategies for enhancing user satisfaction in these critical areas.

While the Gaming category is undeniably important, the decision to exclude it from the current analysis is based on the need for a focused and balanced study on functional, utility-based, and entertainment app experiences. This approach ensures that the insights derived are robust and applicable across a wide range of app types, providing meaningful guidance for improving user satisfaction and engagement.

### Dataset Variables
The dataset contains 11 variables, offering a multifaceted view of user interactions and app attributes. 

- **id**: Unique identifier for each review.
- **app_id**: Unique identifier for the app being reviewed.
- **app_name**: Name of the mobile application being reviewed.
- **category_id**: Four-digit identifier representing the category or genre of the app.
- **category**: Category or genre name of the app.
- **author**: Name or identifier of the reviewer.
- **rating**: Numeric rating provided by the author for the app.
- **content**: Detailed content of the review provided by the author.
- **vote_sum**: Total sum of votes on the usefulness of the rating.
- **vote_count**: Number of votes on the usefulness of the rating.
- **date**: Date when the review was written.

These variables collectively evince the mobile app user experience and provide a basis for analyzing user feedback and extracting actionable intelligence about areas for improvement, feature requests, unmet needs, and user preferences.
